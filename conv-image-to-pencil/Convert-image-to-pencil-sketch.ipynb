```python
import cv2
```

```python
original_image = cv2.imread("jerry-image.png")
```

```python
gray_image = cv2.cvtColor(original_image,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite("jerry(sketch).png",gray_image)
```

```python
inverted_image = 255-gray_image
```

```python
cv2.imwrite("jerry(inverted).png",inverted_image)
```

```python
blurred = cv2.GaussianBlur(inverted_image,(21,21),0)
cv2.imwrite("jerry(sketch).png",blurred)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch =cv2.divide(gray_image,inverted_blurred,scale=256.0)
cv2.imwrite("jerrry-pencil-sketch.png",pencil_sketch)
```
